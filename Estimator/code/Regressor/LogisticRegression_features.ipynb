```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
```

```python
X_train_1 = np.load("../../../Processing/data/X_train_1.npy")
y_train_1 = np.load("../../../Processing/data/y_train_1.npy")
X_train_2 = np.load("../../../Processing/data/X_train_2.npy")
y_train_2 = np.load("../../../Processing/data/y_train_2.npy")
X_test_1 = np.load("../../../Processing/data/X_test_1.npy")
real_test = np.load("../../../Processing/data/real_test.npy")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import StratifiedKFold
parameter_grid ={
                 "C":np.arange(0.1, 2, 0.1)
                 }
cross_validation = StratifiedKFold(y_train_1, n_folds=5)

clf = GridSearchCV(estimator = LogisticRegression(solver ="lbfgs", penalty='l2', multi_class="multinomial"),
                   param_grid = parameter_grid,
                  cv = cross_validation)
```

```python
clf = clf.fit(X_train_1, y_train_1)
print(clf.score(X_train_1, y_train_1))
```

```python
output = clf.predict_proba(X_train_2)[:, 1]
np.save("../../data/lr_features.npy", output)
```

```python
output = clf.predict_proba(X_test_1)[:, 1]
np.save("../../data/lr_features_test.npy", output)
```

```python
output = clf.predict_proba(real_test)[:, 1]
np.save("../../data/lr_features_real_test.npy", output)
```
