```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
```

```python
combined = pd.read_csv("../data/combined.csv")
```

```python
combined.head()
```

valuesを標準化

```python
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
combined_std = stdsc.fit_transform(combined)
```

combinedを分割し、targetsを求める関数を定義

```python
def recover_train_test_target():
    global combined
    
    train0 = pd.read_csv('../../トレーニングデータ+テストデータ/train.csv')
    
    targets = train0.Survived
    train = combined.ix[0:890]
    test = combined.ix[891:]
    
    return train,test,targets
```

```python
train,test,targets = recover_train_test_target()
```

```python
train
```

ExtraTreesClassifierを用いて特徴量を削減

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
clf = ExtraTreesClassifier(n_estimators=200)
clf = clf.fit(train, targets)
```

```python
features = pd.DataFrame()
features['feature'] = train.columns
features['importance'] = clf.feature_importances_
```

```python
features.sort(['importance'],ascending=False)
```

SelectFromModelを用いて有用な特徴量のみ取り出す

```python
model = SelectFromModel(clf, prefit=True)
all_train = model.transform(train)
all_train.shape
```

```python
real_test = model.transform(test)
real_test.shape
```

all_trainをX_train, y_train,に分割し、targetsからX_test_1tとy_test_1を作成

```python
from sklearn.cross_validation import train_test_split
X_train, X_test_1, y_train, y_test_1 = train_test_split(all_train, targets.values, test_size=0.33, random_state=0)
```

X_trainをX_train_1,X_train_2に分割し、y_trainからy_train_1,y_train_2

```python
X_train_1, X_train_2, y_train_1, y_train_2 = train_test_split(X_train, y_train, test_size=0.5, random_state=0)
```

```python
np.save('../data/X_train_1.npy', X_train_1)
np.save('../data/y_train_1.npy', y_train_1)
np.save('../data/X_train_2.npy', X_train_2)
np.save('../data/y_train_2.npy', y_train_2)
np.save('../data/X_test_1.npy', X_test_1)
np.save('../data/y_test_1.npy', y_test_1)
```
