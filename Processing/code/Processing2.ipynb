```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
```

```python
combined = pd.read_csv("../data/combined.csv")
```

```python
combined.head()
```

valuesを標準化

```python
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
combined_std = stdsc.fit_transform(combined)
combined_std = pd.DataFrame(combined_std, columns = combined.columns)
combined_std.head()
combined=combined_std
combined
```

combinedを分割し、targetsを求める関数を定義

```python
def recover_train_test_target():
    global combined
    
    train0 = pd.read_csv('../../Training_test_data/train.csv')
    
    targets = train0.Survived
    train = combined.ix[0:890]
    test = combined.ix[891:]
    
    return train,test,targets
```

```python
train,test,targets = recover_train_test_target()
```

```python
train.head()
```

SelectFromModelを用いて特徴量を削減

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
clf = ExtraTreesClassifier(n_estimators=200)
clf = clf.fit(train, targets)
```

```python
features = pd.DataFrame()
features['feature'] = train.columns
features['importance'] = clf.feature_importances_
```

```python
features.sort(['importance'],ascending=False)
```

SelectFromModelを用いて有用な特徴量のみ取り出す

```python
model = SelectFromModel(clf, prefit=True)
all_train = model.transform(train)
all_train.shape
```

```python
real_test = model.transform(test)
real_test.shape
```

all_trainをX_train, y_trainに分割し、targetsからX_testとy_testを作成

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(all_train, targets.values, test_size=0.15, random_state=0)
```

```python
np.save('../data/X_train.npy', X_train)
np.save('../data/y_train.npy', y_train)
np.save('../data/X_test.npy', X_test)
np.save('../data/y_test.npy', y_test)
np.save('../data/real_test.npy', real_test)
np.save('../data/targets.npy', targets)
np.save('../data/all_train.npy', all_train)
```
